```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/df_file.csv")
df.head()
```

```python
print("Shape:", df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
df["Label"].value_counts().plot(kind="bar")
plt.title("Phân phối nhãn")
plt.xlabel("Label")
plt.ylabel("Số lượng")
plt.show()
```

```python
df["text_len"] = df["Text"].str.split().apply(len)
df["text_len"].hist(bins=50)
plt.title("Độ dài văn bản (số từ)")
plt.xlabel("Số từ")
plt.ylabel("Số lượng")
plt.show()

df["text_len"].describe()
```

```python
duplicates = df.duplicated(subset=["Text"]).sum()
print("Số văn bản trùng lặp:", duplicates)
```
