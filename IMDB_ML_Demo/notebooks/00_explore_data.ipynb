```python
# Cell 1: Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Cell 2: Load data
df = pd.read_csv("../data/raw/IMBD.csv")
df.head()
```

```python
# Cell 3: Info & nulls
df.info()
df.isna().sum().sort_values(ascending=False)
```

```python
# Cell 4: Thống kê rating, genre
sns.histplot(df["rating"].dropna(), bins=20)
plt.title("Phân phối rating")
```

```python
df["genre"].value_counts().head(15).plot(kind="barh")
plt.title("Top 15 thể loại phổ biến")
```

```python
# Cell 5: Tương quan votes - rating
sns.scatterplot(data=df, x="votes", y="rating")
plt.xscale("log")
plt.title("Mối quan hệ giữa votes và rating")
```
